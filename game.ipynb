```python
class person:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self,amt):
        self.balance+=amt
    def withdraw(self,amt):
        if amt>self.balance:
            return -1
        else:
            self.balance-=amt
            return amt
    def __str__(self):
        return f"hello {self.name}, you have an amount of {self.balance} to bet from"
```
